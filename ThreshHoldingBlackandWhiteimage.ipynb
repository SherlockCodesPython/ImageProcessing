```python
import matplotlib.pyplot as plt
import skimage
from skimage import data , io,color
from skimage.filters import threshold_otsu, threshold_local, try_all_threshold


#img1 = (skimage.io.imread(r"C:\Users\Anik Chatterjee\AppData\Local\Programs\Python\Python37-32\Lib\site-packages\skimage\data\motorcycle_right.png"))
img1 = color.rgb2gray(skimage.io.imread(r"C:\Users\Anik Chatterjee\AppData\Local\Programs\Python\Python37-32\Lib\site-packages\skimage\data\motorcycle_right.png"))
plt.figure(figsize=(8,8))
plt.imshow(img1, cmap='gray')

img1 - img1 *255
thresh = threshold_otsu(img1 )
#thresh = threshold_local(img1,block_size=25)
print('Threshold value of this image is ', thresh)
img1_binary = img1 > thresh
print('binary values of the image',img1_binary)
fig , axes = plt.subplots(1,2, figsize=(14,8),sharex = True , sharey=True)
ax= axes.ravel()

ax[0].imshow(img1,cmap='gray')
ax[0].set_title=('Original Image')
ax[0].axis('off')

ax[1].imshow(img1_binary, cmap='gray')
ax[1].set_title=('Black and white image')
ax[1].axis('off')

fig.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(6,6))
ax.hist(img1.ravel())
ax.set_title('Histogram')
ax.set_xlabel('Pixel Intensity')
ax.set_ylabel('Pixel count')
ax.axvline(thresh,color='y')

#Try all thresholds
fig, ax = try_all_threshold(img1, figsize =(16,12),verbose=False)
plt.show()

```
