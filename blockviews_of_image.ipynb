```python
import numpy as np
from matplotlib import pyplot as plt
import skimage.io
from skimage import color
from skimage.util import view_as_blocks



#image conversion pnj to jpg
from PIL import Image
im1 = Image.open(r'C:\Users\Anik Chatterjee\AppData\Local\Programs\Python\Python37-32\Lib\site-packages\skimage\data\astronaut.png')
im1.save(r'C:\Users\Anik Chatterjee\AppData\Local\Programs\Python\Python37-32\Lib\site-packages\skimage\data\astronaut.jpg')

three_dogs = skimage.io.imread(r"C:\Users\Anik Chatterjee\AppData\Local\Programs\Python\Python37-32\Lib\site-packages\skimage\data\astronaut.jpg")
three_dogs = color.rgb2gray(three_dogs)
plt.imshow(three_dogs, cmap="gray")


three_dogs.shape

block_shape=(4,4)
three_dogs_blocks = view_as_blocks(three_dogs,block_shape)

three_dogs_blocks.shape
#here we are reducing the dimensions of the image in the block view i.e from (128, 128, 4, 4) to i-d array (128, 128, 16)
flattened_blocks = three_dogs_blocks.reshape(three_dogs_blocks.shape[0],three_dogs_blocks.shape[1],-1)
print("shape of the blocks image:", three_dogs_blocks.shape)
print("shape of the flattened image :", flattened_blocks.shape)
# 4/4 block is reduced to 16

#Pooling values for the image
#mean pooling or avarage pooling
mean_blocks = np.mean(flattened_blocks, axis = 2)
plt.imshow(mean_blocks, interpolation = 'nearest', cmap = 'gray')
#image is blured due to mean pooling
#WE have extracted mean pooling 
max_blocks = np.max(flattened_blocks, axis = 2)
plt.imshow(max_blocks, interpolation = 'nearest', cmap = 'gray')
#extracts most prominent features of the image

median_blocks = np.median(flattened_blocks, axis = 2)
plt.imshow(median_blocks, interpolation = 'nearest', cmap = 'gray')
```
